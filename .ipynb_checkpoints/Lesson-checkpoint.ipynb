```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("coffees.csv")
```

```python
data[data.coffees.isnull()]
```

```python
data.coffees = pd.to_numeric(data.coffees, errors="coerce")

data.head()
```

```python
data.dropna(inplace=True)

data.head()
```

```python
data.coffees = data.coffees.astype(int)
```

```python
data.timestamp = pd.to_datetime(data.timestamp)

data.dtypes
```

```python
data.describe(include="all")
```

```python
data.tail(n=10)
```

```python
data = data[data.timestamp < "2013-03-01"]
data.tail()
```

```python
plt.figure(figsize=(15,4))
plt.plot(data.timestamp, data.coffees, ".-")
plt.xlabel('timestamp')
plt.ylabel('coffees')
plt.legend(['Coffees'])
plt.show()
```

```python
data.contributor.value_counts().plot(kind="bar")
```

```python
weekdays = data.timestamp.dt.weekday

data = data.assign(weekdays=weekdays)

data.head()
```

```python
weekday_names = ["Monday", "Tuesday", "Wednesday", "Thirsday", "Friday", "Saturday", "Sunday"]

weekday_dict = {key: weekday_names[key] for key in range(7)}

def days_of_week(idx):
    return weekday_dict[idx]

data.weekdays = data.weekdays.apply(days_of_week)

data.head()
```
